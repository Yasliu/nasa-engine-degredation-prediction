# Info

Here I will note down the contents that seem useful for this dataset to learn it properly.

## Understanding the columns:
- `unit number` --> This is basically the engine that we are currently focusing on, there will be `multiple unit numbers` each notifying a different engine.
- `time, in cycles` --> This determines a full flight cycle, each increment in value means one more flight (TakeOff, Cruise, Landing) 
- `operational setting 1` --> Altitude (How high the plane is flying)
- `operational setting 2` --> Mach Number (Speed of aircraft relative to the speed of sound.)
- `operational setting 3` --> Throttle Resolver Angle (TRA) (How far down the pilot has pushed down the throttle lever to command engine power.)

## Feature Engineering

First thing we plan to do now is calculate the rolling averages.

- Firstly we get ourselves a rolling avg of window = 20 to provide us with the average drop of RUL over a longer period of time
- Secondly, we get the same but for window = 5 which shows a more immediate drop. 
- Then we subtract both to calculate the divergance with a certain `threshold` which will allow us to determine if an engine is degrading fast / slow and whether its degrading or not. [ Note --> In case of confusion, the ML Model you choose will do this because it will be trained with the y_train which will be the Piecewise_RUL]

- Then we make a column which is going to be the Trend_Acceleration -- original - shift(5) which shows the model regarding the stability as two engines, one new and one old can be idling at 900 and 950 degrees respectively -- this tracks sudden spikes.

- Then we engineer a systemic Volatility column where it basically finds the rolling std for a window of 5 which basically would show that if there is a steady increase in value then a component is breaking apart or losening.

- Fleet-Wide Cross-Sensor Ratio --> this will be the strongest negatively corr / strongest positively corr. so as the engine will degrade, the top value will increase while lower value decreases causing the value to skyrocket.

## Eval Vs Objective

Understand that these two are different. You create an objective function for your model itself so that it can push itself in the right direction during training. In general, XGBoost uses an MSE in order to push itself in the right direction whereas you want the Nasa's custom penalty function to do well in this. So, as a result you end up writing a custom objective function for it.

Now , you trained the model and you ran it. How do you know how the evaluation went? Using MSE? Nope. Then what was the point of it all. So, now you write the eval function which in a sense would be similar to that of the objective but in the end returning the score that is human-readable so that we can understand how it actually went when we tested it.
